```python
rootdir_glob = 'unzipped/**/*' # Note the added asterisks
# This will return absolute paths
file_list = [f for f in iglob(rootdir_glob, recursive=True) if zipfile.is_zipfile(f)]
file_list
for file in file_list:
    file = file.split("/", -1)
```

```python
file_list[0]
```

```python
new_list = []
for file in file_list:
    file = file.split("/w")
    #file = file[0].split("ed/")
    new_list.append(file[0])
```

```python
name_list = []
for file in new_list:
    file = file.split("ed/")
    file = file[1].split("/annotation/")
    file= file[0]+'_A_'+file[1]
    name_list.append(file)
name_list
```

```python
os.makedirs('unzipped/extracted/xmi')
os.makedirs('unzipped/extracted/xml')
```

```python
n = 0
for i,ii in zip(new_list,name_list):

    zipfilename = os.listdir(i)
    if len(zipfilename)==1:
        with ZipFile(i+"/"+zipfilename[0],'r') as zipObj:
            zipObj.extractall(path=i+"/")
    unzipped_file = open(i+"/admin.xmi")
    new_text = open("unzipped/extracted/xmi/"+ii,"w")
    typesystem = open(i+"/TypeSystem.xml")
    new_type = open("unzipped/extracted/xml/"+ii+"typesystem.xml","w")
    for j in unzipped_file:
        new_text.write(j)
    new_text.close()
    
    for t in typesystem:
        new_type.write(t)
    new_type.close()
    n += 1
```
